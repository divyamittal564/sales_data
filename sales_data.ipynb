```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.random.randint(20,100,30)
print(data)
```

```python
total_sales=data.sum()
print("Total sales:",total_sales)
```

```python
avg=data.mean()
print("Average:",avg)
```

```python
highest_sales=data.max()
highest_index=data.argmax()
print("\n Highest sales:",highest_sales)
print("\n Highest index:",highest_index+1)

lowest_sales=data.min()
lowest_index=data.argmin()
print("\n Lowest sales:",lowest_sales)
print("\n Lowest index:",lowest_index+1)
```

```python
profit_loss_percent = ((data[1:] - data[:-1]) / data[:-1]) * 100
print("\nDaily Profit/Loss % (Day 2 to Day 30):\n", profit_loss_percent)
```

```python
plt.plot(data, marker='o')
plt.title("Daily Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales Amount")
plt.grid(True)
plt.show()
```
